```python
from ultralytics import YOLO
import cv2
import cvzone

# Load video file
cap = cv2.VideoCapture(r'Assets\bottle.mp4')

# Set the video height and width (optional, for specific resolution)
cap.set(3, 640)  # Width
cap.set(4, 480)  # Height

# Load YOLO model
model = YOLO(r'model.pt')

# Show the video
while True:
    ret, img = cap.read()
    if not ret:  # Break the loop if the video ends
        break

    # Process the frame with YOLO model
    results = model(img, stream=True)

    # Loop through the results and draw bounding boxes with labels
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            conf, cls = box.conf[0], box.cls[0]
            w, h = x2 - x1, y2 - y1

            # Draw a fancy rectangle around the detected object
            cvzone.cornerRect(img, (int(x1), int(y1), int(w), int(h)), 20, 5, rt=2)

            # Add the label and confidence score
            cvzone.putTextRect(
                img,
                f'{model.names[int(cls)]} {conf:.2f}',
                (int(x1), max(int(y1), 0)),
                1,
                2,
                border=2,
            )

    # Resize the frame to a manageable size (e.g., 640x480)
    img_resized = cv2.resize(img, (1000, 780))

    # Show the resized frame
    cv2.imshow('Video Frame', img_resized)

    # Break the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import cvzone
from sort import *
import numpy as np

# Load video file
cap = cv2.VideoCapture('Assets/bottle.mp4')
mask = cv2.imread('Assets/mask.png')
cap.set(3, 640)  # Width
cap.set(4, 480)  # Height

# Load YOLO model
model = YOLO('model.pt')

# Tracking
tracker = Sort(max_age=20, min_hits=3, iou_threshold=0.3)
limits = [390, 500, 430, 570]
counter = []

# Initialize video writer variables
out = None
frame_width, frame_height = None, None

while True:
    ret, img = cap.read()
    if not ret:  # Break the loop if the video ends
        break

    # Determine frame dimensions (initialize only once)
    if frame_width is None or frame_height is None:
        frame_height, frame_width = img.shape[:2]
        out = cv2.VideoWriter(
            'output.mp4',
            cv2.VideoWriter_fourcc(*'mp4v'),
            20.0,
            (frame_width, frame_height)
        )

    # Process the frame with YOLO model
    #load the grpahic image in the img
    grapic = cv2.imread('Assets/graphic.png' , cv2.IMREAD_UNCHANGED)
    img = cvzone.overlayPNG(img, grapic, (0, 0))
    img_region = cv2.bitwise_and(img, mask)
    detections = np.empty((0, 5))
    results = model(img_region, stream=True)

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            conf, cls = box.conf[0], box.cls[0]
            w, h = x2 - x1, y2 - y1

            cls_name = model.names[int(cls)]
            if cls_name == 'bottle':
                currentArray = np.array([x1, y1, x2, y2, conf])
                detections = np.vstack((detections, currentArray))

    resultsTracker = tracker.update(detections)

    cx, cy = (limits[0] + limits[2]) // 2, ((limits[1] + limits[3]) // 2) - 20
    cv2.circle(img, (cx, cy), 10, (0, 255, 0), -1)

    for track in resultsTracker:
        x1, y1, x2, y2, id = track
        if x1 < cx < x2 and y1 < cy < y2:
            if id not in counter:
                counter.append(id)

        cvzone.putTextRect(img, f'{len(counter)}', (150, 140), 3, 3, offset=5)
        cvzone.cornerRect(img, (int(x1), int(y1), int(x2 - x1), int(y2 - y1)), 20, 5, rt=2)
        cvzone.putTextRect(img, f'Filled Bottle: {int(id)}', (int(x1), int(y1)), 1, 1, offset=5)

    # Write the frame to the output video
    out.write(img)

    # Display the frame
    cv2.imshow('Video Frame', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
if out:
    out.release()
cv2.destroyAllWindows()
```
